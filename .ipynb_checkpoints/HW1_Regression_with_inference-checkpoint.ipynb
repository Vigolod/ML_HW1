* <font color='#9933CC'>Вместо удаления признак `torque` разделите на два: собственно `torque` и `max_torque_rpm`.</font> Учтите единицы измерения. Они разные ☹

```python
df_train['mileage'].str.extract(r'([^\d\s\.,]+)', expand=False).unique()
```

```python
df_train['torque'].str.extract(r'([a-zA-Z]+)(?:@|at)', expand=False).unique()
```

```python
def parse_features(df):
    df = df.copy()
    kmkg_mask = df['mileage'].str.contains('km/kg', na=False)
    
    for column in ['mileage', 'engine', 'max_power']:
        df[column] = pd.to_numeric(df[column].str.extract(r'(\d+\.?\d*)', expand=False))
    df.loc[kmkg_mask, 'mileage'] = df.loc[kmkg_mask, 'mileage'] * 0.8
    
    transl = str.maketrans('', '', ',')
    
    df['max_torque_rpm'] = df['torque'].str.extract('(?:@|at)\s*([-\d,\.]+)', expand=False) \
                                                   .str.translate(transl)
    is_range = df['max_torque_rpm'].str.contains('-', na=False)

    first_term = pd.to_numeric(
        df.loc[is_range, 'max_torque_rpm'] \
          .str.extract(r'([\d,\.]+)', expand = False)
    )

    second_term = pd.to_numeric(
        df.loc[is_range, 'max_torque_rpm'] \
          .str.extract(r'-(\d+)', expand = False)
    )

    df.loc[is_range, 'max_torque_rpm'] = (first_term + second_term) / 2
    df['max_torque_rpm'] = pd.to_numeric(df['max_torque_rpm'])
    
    kgm_mask = df['torque'].str.lower().str.contains('kgm', na=False)
    
    df['torque'] = pd.to_numeric(df['torque'].str.extract(r'\s*([\d,\.]+)', expand=False) \
                                                 .str.translate(transl))
    df.loc[kgm_mask, 'torque'] = df.loc[kgm_mask, 'torque'] * 9.80665
    return df
```

```python
df_train = parse_features(df_train)
df_test = parse_features(df_test)
```

(0.5 балла) <font color='#9933CC'>Закодируйте категориалльные фичи и ``seats`` методом OneHot-кодирования</font>

> Обратите внимание, что во избежание мультиколлинеарности следует избавиться от одного из полученных столбцов при кодировании каждого признака методом OneHot. 

```python
from sklearn.base import TransformerMixin

class CustomTrasnformer(TransformerMixin):
    def __init__(self, cols=None, transformer=None):
        self.cols = cols
        self.transformer = transformer or StandardScaler()
        
    def fit(self, X, y=None):
        if self.cols is None:
            self.cols = X.columns
        num_cols = X.copy()[self.cols]
        self.transformer.fit(num_cols, y)
        return self
    def transform(self, X, y=None):
        X_res = X.copy()
        num_cols_tr = self.transformer.transform(X_res[self.cols])
        for i, col in enumerate(self.cols):
            X_res[col] = num_cols_tr[:,i]
        return X_res
    def set_params(self, **params):
        self.transformer.set_params(**params)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder # или можно использовать get_dummies из библиотеки pandas


scale_ohe = Pipeline([
    (
        'scaler',
        CustomTrasnformer(
            cols=df_train.select_dtypes(['int', 'float'])  \
                         .drop(['selling_price', 'seats'], axis=1) \
                         .columns
        )
    ),
    (
        'ohe',
        ColumnTransformer(
            [
                (
                    'one_hot_encoder', 
                    OneHotEncoder(drop='first', sparse=False), 
                    np.append(X_train_cat.columns[X_train_cat.dtypes == 'object'], 'seats')
                )
            ]
        )
    )
])
scale_ohe.fit(X_train_cat)
X_train_ohe = scale_ohe.transform(X_train_cat)
X_test_ohe = scale_ohe.transform(X_test_cat)
```

```python
df_train = pd.read_csv('https://raw.githubusercontent.com/Murcha1990/MLDS_ML_2022/main/Hometasks/HT1/cars_train.csv')
df_test = pd.read_csv('https://raw.githubusercontent.com/Murcha1990/MLDS_ML_2022/main/Hometasks/HT1/cars_test.csv')
df_train = df_train[~df_train.drop('selling_price', axis=1).duplicated()]
df_train = parse_features(df_train)
df_test = parse_features(df_test)
#df_train = df_train[df_train['selling_price'] < df_train['selling_price'].quantile(0.95)]
```

```python
for df in [df_train, df_test]:
    df['name'] = df['name'].str.extract(r'\s*([a-zA-Z]+)\s', expand=False)
    df.drop(['torque'], axis=1, inplace=True)
    df['max_torque_rpm'] = (df['max_torque_rpm'] < 2500).astype('int')
    
    df['km_driven'] = np.log(df['km_driven'] + 1e-10)
    df.drop('mileage', axis=1, inplace=True)

    df.loc[(df['owner'] == 'Third Owner') | 
             (df['owner'] == 'Fourth & Above Owner'),
            'owner'] = 'Third & Above Owner'

    df.loc[df['owner'] == 'Test Drive Car', 'owner'] = 'First Owner'
    
    df['year_sq'] = df['year'] ** 2
    nan_cols = df.columns[df.isna().sum() > 0]
    for col in nan_cols:
        df[col+'_was_nan'] = df[col].isna().astype('int')
```

Заказчик просил вас посчитать кастомную метрику -- среди всех предсказанных цен на авто посчитать долю предиктов, отличающихся от реальных цен на эти авто не более чем на 10% (в одну или другую сторону)

<font color='#9933CC'>Сделайте это самостоятельно для лучшей из своих моделей</font>

```python
def business_metrics(y_true, y_pred):
    gap = y_true * 0.1
    score = (np.abs(y_true - y_pred) <= gap).sum() / y_true.shape[0]
    return score

business_metrics(y_test, best_ridge.predict(X_test))
```
